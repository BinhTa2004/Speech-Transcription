```python
import whisper
import sounddevice as sd
import numpy as np
import tempfile
import os
import soundfile as sf
import uuid
import datetime

# Load model Whisper (cÃ³ thá»ƒ dÃ¹ng 'tiny', 'base', 'small', 'medium', 'large')
model = whisper.load_model("base")

# Cáº¥u hÃ¬nh
SAMPLE_RATE = 16000
CHUNK_DURATION = 5  # giÃ¢y
CHUNK_SAMPLES = SAMPLE_RATE * CHUNK_DURATION
```

```python
def record_and_transcribe():
    print("ğŸ™ Báº¯t Ä‘áº§u ghi Ã¢m vÃ  nháº­n diá»‡n (nháº¥n Ctrl+C Ä‘á»ƒ dá»«ng)...")

    save_dir = "recordings"
    os.makedirs(save_dir, exist_ok=True)

    try:
        while True:
            print("\n--- Ghi Ã¢m ---")
            audio = sd.rec(CHUNK_SAMPLES, samplerate=SAMPLE_RATE, channels=1, dtype=np.float32)
            sd.wait()

            # Äáº·t tÃªn file theo timestamp
            filename = datetime.datetime.now().strftime("%Y%m%d_%H%M%S") + ".wav"
            save_path = os.path.join(save_dir, filename)

            # LÆ°u file WAV tháº­t
            sf.write(save_path, audio, SAMPLE_RATE)
            print(f"ğŸ’¾ ÄÃ£ lÆ°u file: {save_path}")

            # Cháº¡y Whisper
            result = model.transcribe(save_path, language="vi")
            print("ğŸ“ Káº¿t quáº£:", result["text"])

            # á» Ä‘Ã¢y cÃ³ thá»ƒ gá»i luÃ´n hÃ m mÃ´ táº£ audio (audio captioning) báº±ng save_path
            # description = audio_caption_model.describe(save_path)
            # print("ğŸ“œ MÃ´ táº£:", description)

    except KeyboardInterrupt:
        print("\nğŸ›‘ Dá»«ng.")

if __name__ == "__main__":
    record_and_transcribe()
```
